#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fitbit_2_average_steps_month = fitbit_df[['Months_encoded', 'Steps']].groupby('Months_encoded', as_index=False).mean()
average_weekend_work = fitbit_df[['Work_or_Weekend', 'Steps']].groupby('Work_or_Weekend', as_index=False).mean()
average_weekend_work['Work_or_Weekend'] = np.where(average_weekend_work['Work_or_Weekend']==1,'Work','Weekend')
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(12, 4))
axes.set(xlabel= 'MONTH', 
         ylabel= 'STEPS',
         title = 'MEDIAN STEPS BY MONTH_ENCODED')
axes.bar(fitbit_2_average_steps_month['Months_encoded'], fitbit_2_average_steps_month['Steps'])
plt.show()


# MEDIAN STEPS BY WORK_OR_WEEKEND
fig, axes = plt.subplots(ncols=1, nrows=1, figsize=(8, 4)) 
axes.set(ylim=[10000, 10200],
         xlabel='WORK_OR_WEEKEND', 
         ylabel='Steps', 
         title='MEDIAN STEPS BY WORK_OR_WEEKEND')
axes.bar(average_weekend_work['Work_or_Weekend'], average_weekend_work['Steps'])
plt.show()

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
columns = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
```

#### Plot the correlation matrix using seaborn.

```python
corr = titanic_df.corr()
ax = sns.heatmap(corr, 
                 vmin=-1, vmax=1, center=0,
                 cmap=sns.diverging_palette(10, 220, n=200),
                 square=True)
ax.set_xticklabels(ax.get_xticklabels(),
                   rotation=45,
                   horizontalalignment='right');
```
